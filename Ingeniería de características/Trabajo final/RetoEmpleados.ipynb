1. Se importa el archivo CSV con la información a analizar y se guarda en un DataFrame:

```python
import pandas as pd

EmpleadosAttrition = pd.read_csv('/content/empleadosRETO.csv')
EmpleadosAttrition
```

----

2. Se eliminan posibles columnas innecesarias para el análisis (EmployeeCount, EmployeeNumber, Over18, StandardHours):

```python
EmpleadosAttrition = EmpleadosAttrition.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1)
EmpleadosAttrition
```

---------

3. Se detectó que no existe una columna con los años que el empleado tiene trabajando en la compañia, se procede a crearla y tendrá el nombre de **WorkingYearsAttrition**. Existe una columna llamada **TotalWorkingYears**, que indica su experiencia total, pero no únicamente en la compañía:

```python
from datetime import date

fechaHoy = date.today()
print("La fecha que se tomará en cuenta para el cálculo de años es: ", fechaHoy)

EmpleadosAttrition['WorkingYearsAttrition'] = fechaHoy.year  - (EmpleadosAttrition['HiringDate'].str.split(pat='/').str[2].astype(int)) 
EmpleadosAttrition
```

--------

4. Se crea una columna llamada **Year** que es el año de contratación del empleado:

```python
EmpleadosAttrition['Year'] = EmpleadosAttrition['HiringDate'].str.split(pat='/').str[2].astype(int)
EmpleadosAttrition.info()
EmpleadosAttrition
```

-------

5. Se crea una columna llamada **YearsAtCompany** que son los años que lleva el empleado en la compañía hasta 2018:

```python
anio = 2018
EmpleadosAttrition['YearsAtCompany'] = anio  - EmpleadosAttrition['Year'] 
EmpleadosAttrition
```

------

6. Se depura la columna **DistanceFromHome**, quitando el texto y dejando únicamente la parte numérica. :

```python
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome'].str.split(pat=' ').str[0].astype(int)
EmpleadosAttrition
```

---------------

7. Se renombra la columna **DistanceFromHome** a **DistanceFromHome_km**:

```python
EmpleadosAttrition.rename(columns = {'DistanceFromHome':'DistanceFromHome_km'}, inplace = True)
EmpleadosAttrition
```

--------

8. Se crea una nueva columna llamada **DistanceFromHome** y se le asignan los valores números de la columna **DistanceFromHome_km**:

```python
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome_km']
EmpleadosAttrition
```

------

9. Se depura el DataFrame con las columnas que ya no se utilizarán (Year, HiringDate, DistanceFromHome_km:

```python
EmpleadosAttrition.drop(['Year', 'HiringDate', 'DistanceFromHome_km'], inplace=True, axis=1)
EmpleadosAttrition
```

------

10. Se genera un nuevo Frame que contiene el sueldo promedio por departamento:

```python
SueldoPromedioDepto = EmpleadosAttrition.groupby(['Department'])[['MonthlyIncome']].mean().reset_index()
SueldoPromedioDepto
```

-------

11. Se escala la variable **MonthlyIncome** para que tenga un valor entre 0 y 1 (Escalamiento min-max):

```python
from sklearn import preprocessing

escalador = preprocessing.MinMaxScaler()
frameEscalado = escalador.fit_transform(EmpleadosAttrition[['MonthlyIncome']])
EmpleadosAttrition['MonthlyIncome'] = frameEscalado
print("Columna MonthlyIncome escalada: ")
EmpleadosAttrition['MonthlyIncome']
```

---------

12. Se convierten las variables categóricas en variables númericas (BusinessTravel, Department, EducationField, Gender, JobRole, MaritalStatus, Attrition):

```python
dummies = pd.get_dummies(EmpleadosAttrition)
dummies
```

---------------

13. Se realiza la evaluación de variables a través de correlación lineal con respecto a Attrition. Las variables que se consideran en el análisis serán aquellas con un valor mayor a 0.1 y se guardarán en el frame **EmpleadosAttritionFinal**:

```python
correlacion = dummies.corr()
correlacion = correlacion.where(correlacion.abs() > 0.1)

correlacion = correlacion.loc[['Attrition_No', 'Attrition_Yes']]
correlacion = correlacion.dropna(axis=1, how='all')
print(correlacion.columns)
print()

EmpleadosAttritionFinal = dummies[ correlacion.columns ]
print(EmpleadosAttritionFinal)
```

-----

14. Se aplica el proceso PCA al frame **EmpleadosAttritionFinal**:

```python
from sklearn.decomposition import PCA

EmpleadosAttritionPCA = PCA(len(EmpleadosAttritionFinal.columns))
EmpleadosAttritionPCA.fit(EmpleadosAttritionFinal)

print("Componentes: ")
print(EmpleadosAttritionPCA.components_)
print()

print("Evaluación de los componentes: ")
print(EmpleadosAttritionPCA.explained_variance_)
print()
```

------

15. Se agregan los componentes principales del proceso PCA a **EmpleadosAttritionFinal** y se agrega a un archivo CSV:

```python
EmpleadosAttritionFinal["C0"] = pd.Series(EmpleadosAttritionPCA.components_[:, 0])
EmpleadosAttritionFinal["C1"] = pd.Series(EmpleadosAttritionPCA.components_[:, 1])
EmpleadosAttritionFinal["C2"] = pd.Series(EmpleadosAttritionPCA.components_[:, 2])
EmpleadosAttritionFinal["C3"] = pd.Series(EmpleadosAttritionPCA.components_[:, 3])
EmpleadosAttritionFinal["C4"] = pd.Series(EmpleadosAttritionPCA.components_[:, 4])
EmpleadosAttritionFinal["C5"] = pd.Series(EmpleadosAttritionPCA.components_[:, 5])
EmpleadosAttritionFinal

from google.colab import files
EmpleadosAttritionFinal.to_csv('EmpleadosAttritionFinal.csv') 
files.download('EmpleadosAttritionFinal.csv')
```
